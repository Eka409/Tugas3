```python
#3
import matplotlib.pyplot as plt

# Data
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des']
kategori_a = [200, 250, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200]
kategori_b = [150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700]
kategori_c = [300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850]

# Membuat Line Chart
plt.figure(figsize=(10, 6))
plt.plot(bulan, kategori_a, marker='o', label='Kategori A', color='blue')
plt.plot(bulan, kategori_b, marker='o', label='Kategori B', color='orange')
plt.plot(bulan, kategori_c, marker='o', label='Kategori C', color='green')

# Menambahkan Judul dan Label
plt.title('Tren Penjualan Produk per Kategori dalam Satu Tahun')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)  # Memutar label bulan untuk keterbacaan
plt.tight_layout()  # Agar semua elemen muat dalam tampilan

# Menampilkan Grafik
plt.show()
```
